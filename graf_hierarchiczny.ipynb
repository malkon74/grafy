```python
!pip install plotly
!pip install matplotlib
!pip install networkx
!pip install pydot

import plotly.graph_objects as go
import matplotlib.pyplot as plt
import networkx as nx

nodes = [
    {'id': 'A', 'label': 'Przygotowanie danych'},
    {'id': 'B', 'label': 'Czyszczenie tekstu'},
    {'id': 'C', 'label': 'Tokenizacja'},
    {'id': 'D', 'label': 'Lematyzacja'},
    {'id': 'E', 'label': 'Embeddingi'},
    {'id': 'F', 'label': 'Modelowanie sekwencji'},
    {'id': 'G', 'label': 'Rozumienie kontekstu'},
    {'id': 'H', 'label': 'Transfer learning'},
    {'id': 'I', 'label': 'Analiza sentymentu'},
    {'id': 'J', 'label': 'Generowanie tekstu'},
    {'id': 'K', 'label': 'Ekstrakcja informacji'},
    {'id': 'L', 'label': 'Zastosowanie i integracja'}
]

edges = [
    {'source': 'A', 'target': 'B'},
    {'source': 'B', 'target': 'C'},
    {'source': 'C', 'target': 'D'},
    {'source': 'D', 'target': 'E'},
    {'source': 'E', 'target': 'F', 'label': 'RNN/LSTM/GRU'},
    {'source': 'E', 'target': 'G', 'label': 'Attention/Transformer'},
    {'source': 'F', 'target': 'H'},
    {'source': 'G', 'target': 'H'},
    {'source': 'H', 'target': 'I'},
    {'source': 'E', 'target': 'J'},
    {'source': 'E', 'target': 'K'},
    {'source': 'E', 'target': 'L'}
]

# Próba z plotly (zapis do pliku HTML)
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=[node['label'] for node in nodes],
        color=["blue" for node in nodes]
    ),
    link=dict(
        source=[edges[i]['source'] for i in range(len(edges))],
        target=[edges[i]['target'] for i in range(len(edges))],
        value=[1 for i in range(len(edges))],
        label=[edges[i].get('label', '') for i in range(len(edges))]
    ))])

fig.update_layout(title_text="Proces NLP", font_size=10)
fig.write_html('wykres_plotly.html')

# Wizualizacja z matplotlib (układ hierarchiczny)
G = nx.DiGraph()
G.add_nodes_from([node['id'] for node in nodes])
G.add_edges_from([(edge['source'], edge['target']) for edge in edges])

labels = {node['id']: node['label'] for node in nodes}
pos = nx.nx_pydot.pydot_layout(G, prog='dot')

plt.figure(figsize=(12, 8))
nx.draw(G, pos, labels=labels, node_size=1500, node_color="skyblue", font_size=10, arrowsize=20)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
plt.title("Proces NLP")
plt.show()
```
